```python
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
# Set up the Selenium WebDriver (ensure you have the appropriate driver installed)
driver = webdriver.Chrome()  # or webdriver.Firefox() if using Firefox
```

```python
# Step 1: Load the webpage using Selenium
driver.get(url)
```

```python
# Step 3: Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python

# Step 4: Find the div with id "block-value-graph"
block_value_graph = soup.find('div', id='block-value-graph')
```

```python
print(block_value_graph)
```

```python
# Step 5: Find the svg element within the block-value-graph div
svg = block_value_graph.find('svg', class_='pd-value-svg')
# Step 6: Extract the data from the svg element
svg_content = svg.prettify()  # Pretty print the SVG content for inspection
print(svg_content)
```

```python
# Step 6: Find all g elements with class "hoverGroup"
hover_groups = svg.find_all('g', class_='hoverGroup')
print(hover_groups)
```

```python
values = []
```

```python
for group in hover_groups:
    hover_date = group.find('text', class_='hoverDate').get_text()
    hover_val = group.find('text', class_='graphVal hoverVal').get_text()
    values.append((hover_date, hover_val))
```

```python
# Print the extracted data
for date, val in values:
    print(f"Date: {date}, Value: {val}")
```

```python
def get_data_for_player():
    url = "https://keeptradecut.com/dynasty-rankings/players/jonathon-brooks-1568?format=1"
    driver = webdriver.Chrome()
    # Step 1: Load the webpage using Selenium
    driver.get(url)
    # Step 3: Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    # Step 4: Find the div with id "block-value-graph"
    block_value_graph = soup.find('div', id='block-value-graph')
    svg = block_value_graph.find('svg', class_='pd-value-svg')
    hover_groups = svg.find_all('g', class_='hoverGroup')
    values = []
    for group in hover_groups:
        hover_date = group.find('text', class_='hoverDate').get_text()
        hover_val = group.find('text', class_='graphVal hoverVal').get_text()
        values.append((hover_date, hover_val))

    print(values)
    return values
```

```python


def main():

    

    data = get_data_for_player()
    print(data)

    for date, val in data:
        print(f"Date: {date}, Value: {val}")



```
